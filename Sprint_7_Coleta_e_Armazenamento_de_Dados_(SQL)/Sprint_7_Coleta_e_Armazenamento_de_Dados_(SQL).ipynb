````python
import pandas as pd
from scipy import stats

# Carregar os dados
df_res_07 = pd.read_csv('/datasets/project_sql_result_07.csv')

# Garantir que o campo 'start_ts' esteja no formato de data
df_res_07['start_ts'] = pd.to_datetime(df_res_07['start_ts'])

# Criar uma coluna para identificar sábados
df_res_07['day_of_week'] = df_res_07['start_ts'].dt.dayofweek
df_res_07['is_saturday'] = df_res_07['day_of_week'] == 5

# Filtrar para sábados chuvosos (condição 'Bad') e sábados não chuvosos (condição 'Good')
saturday_rainy = df_res_07[(df_res_07['is_saturday']) & (df_res_07['weather_conditions'] == 'Bad')]['duration_seconds']
saturday_non_rainy = df_res_07[(df_res_07['is_saturday']) & (df_res_07['weather_conditions'] == 'Good')]['duration_seconds']

# Teste t para amostras independentes
t_stat, p_value = stats.ttest_ind(saturday_rainy, saturday_non_rainy, nan_policy='omit')

# Definir o nível de significância (alpha)
alpha = 0.05

# Resultados
print(f"Valor-p: {p_value}")
if p_value < alpha:
    print("Rejeitamos a hipótese nula. A duração média das viagens nos sábados chuvosos é significativamente diferente.")
else:
    print("Não rejeitamos a hipótese nula. A duração média das viagens nos sábados chuvosos não é significativamente diferente.")
````

``` ~(df_res_07['is_saturday'])).```

Hipóteses:
Hipótese Nula (H₀): A duração média das corridas nos sábados chuvosos é igual à duração média das corridas em outros dias.
Hipótese Alternativa (H₁): A duração média das corridas nos sábados chuvosos é diferente da duração média das corridas em outros dias.
Critério de Teste:
Utilizamos o teste t para amostras independentes, pois estamos comparando as médias de dois grupos independentes (sábados chuvosos e outros dias). Esse teste verifica se as médias dos dois grupos são significativamente diferentes.

Valor-p: Se for menor que 0,05 (nível de significância), rejeitamos a hipótese nula, sugerindo que a chuva no sábado afeta a duração das corridas. Caso contrário, não rejeitamos a hipótese nula.
