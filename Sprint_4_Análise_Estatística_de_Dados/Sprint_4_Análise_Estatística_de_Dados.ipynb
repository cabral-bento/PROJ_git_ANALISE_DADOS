## Agregue os dados por usuário

[Agora, como os dados estão limpos, os agregue por usuário e por período para ter apenas um registro dessas informações. Isso vai facilitar muito as próximas análises.]

```python
# Calcule o número de chamadas feitas por cada usuário por mês. Salve o resultado.
df_calls['calls_per_user'] = df_calls.groupby('user_id')['id'].transform('count')
df_calls.head()
```

```python
#este codigo Calcula o número de chamadas feitas por cada usuário por mês e impreme  reaultado
```

```python
# Calcule a quantidade de minutos gastos por cada usuário por mês. Salve o resultado.
df_calls['minutes_per_user'] = df_calls.groupby('user_id')['duration'].transform('sum')
df_calls.head()
```

```python
#este codigo Calcula a quantidade de minutos gastos por cada usuário por mês e imprime o resultado
```

```python
# Calcule o número de mensagens enviadas por cada usuário por mês. Salve o resultado.
df_messages['messages_per_user'] = df_messages.groupby('user_id')['id'].transform('count')
df_messages.head()
```

```python
#este codigo Calcula o número de mensagens enviadas por cada usuário por mês
```

```python
# Calcule o volume de tráfego de internet usado por cada usuário por mês. Salve o resultado.

df_internet['data_per_user'] = df_internet.groupby('user_id')['mb_used'].transform('sum')
df_internet.head()
```

```python
#este codigo  calcula o volume de tráfego de internet usado por cada usuário por mês
```

[Coloque os dados agregados em um DataFrame para que cada registro nele represente o que cada usuário consumiu em um determinado mês.]

```python
#Junte os dados de chamadas, minutos, mensagens e internet com base em user_id e month

# Primeiro, crie a coluna 'month' se ela ainda não existir
df_calls['month'] = pd.to_datetime(df_calls['call_date']).dt.month
df_messages['month'] = pd.to_datetime(df_messages['message_date']).dt.month
df_internet['month'] = pd.to_datetime(df_internet['session_date']).dt.month

# Agora, agregue os dados por 'user_id' e 'month'
calls_agg = df_calls.groupby(['user_id', 'month']).agg(
    calls_per_user=('id', 'count'),
    minutes_per_user=('duration', 'sum')
).reset_index()

messages_agg = df_messages.groupby(['user_id', 'month']).agg(
    messages_per_user=('id', 'count')
).reset_index()

internet_agg = df_internet.groupby(['user_id', 'month']).agg(
    data_per_user=('mb_used', 'sum')
).reset_index()

# Combine todos os dados com base em 'user_id' e 'month'
df_combined = calls_agg.merge(messages_agg, on=['user_id', 'month'], how='outer')
df_combined = df_combined.merge(internet_agg, on=['user_id', 'month'], how='outer')

# Preenchendo valores ausentes com 0, caso algum usuário não tenha chamadas, mensagens ou uso de internet em um mês
df_combined.fillna(0, inplace=True)

# Visualizando o DataFrame combinado
print(df_combined.head())
```

```python
# Exemplo de df_plans


# Combine todos os dados agregados
df_combined = calls_agg.merge(messages_agg, on=['user_id', 'month'], how='outer')
df_combined = df_combined.merge(internet_agg, on=['user_id', 'month'], how='outer')

# Adicione as informações do plano ao DataFrame combinado
# Primeiro, combine df_users com df_plans para ter uma tabela com user_id, plan e outras informações do plano
df_users_plans = df_users.merge(df_plans, left_on='plan', right_on='plan_name', how='left')

# Agora, adicione as informações do plano ao DataFrame combinado
df_combined = df_combined.merge(df_users_plans[['user_id', 'plan', 'messages_included', 'mb_per_month_included', 'minutes_included', 'usd_monthly_pay', 'usd_per_gb', 'usd_per_message', 'usd_per_minute']], on='user_id', how='left')

# Preenchendo valores ausentes com 0 para as colunas numéricas
df_combined.fillna({'calls_per_user': 0, 'minutes_per_user': 0, 'messages_per_user': 0, 'data_per_user': 0}, inplace=True)

# Visualizando o DataFrame combinado
print(df_combined.head())
df_combined.head()
```

```python
df_combined
```

```python
# Calcule a receita mensal para cada usuário


# Supondo que df_combined já contém as colunas necessárias para o cálculo

# Calcular mensagens extras
df_combined['extra_messages'] = df_combined['messages_per_user'] - df_combined['messages_included']
df_combined['extra_messages'] = df_combined['extra_messages'].clip(lower=0)  # Definir valores negativos como 0

# Calcular minutos extras
df_combined['extra_minutes'] = df_combined['minutes_per_user'] - df_combined['minutes_included']
df_combined['extra_minutes'] = df_combined['extra_minutes'].clip(lower=0)  # Definir valores negativos como 0

# Calcular dados extras
df_combined['extra_data'] = df_combined['data_per_user'] - df_combined['mb_per_month_included']
df_combined['extra_data'] = df_combined['extra_data'].clip(lower=0)  # Definir valores negativos como 0

# Converter extra_data para GB
df_combined['extra_data_gb'] = df_combined['extra_data'] / 1024

# Calcular a receita adicional
df_combined['revenue_messages'] = df_combined['extra_messages'] * df_combined['usd_per_message']
df_combined['revenue_minutes'] = df_combined['extra_minutes'] * df_combined['usd_per_minute']
df_combined['revenue_data'] = df_combined['extra_data_gb'] * df_combined['usd_per_gb']

# Calcular a receita total
df_combined['monthly_revenue'] = df_combined['usd_monthly_pay'] + df_combined['revenue_messages'] + df_combined['revenue_minutes'] + df_combined['revenue_data']

# Exibir as primeiras linhas do DataFrame com a receita calculada
df_combined[['user_id', 'month', 'monthly_revenue']].head()
```

```python
# Compare a duração média das chamadas de cada plano para cada mês. Crie um gráfico de barras para visualizar o resultado.


# Supondo que df_combined já contém a coluna 'plan' e as colunas relacionadas a chamadas
# Primeiro, calcule a duração total das chamadas por plano e mês
calls_per_plan_month = df_combined.groupby(['plan', 'month']).agg(
    total_duration=('minutes_per_user', 'sum'),
    total_calls=('calls_per_user', 'sum')
).reset_index()

# Calcule a duração média das chamadas por plano e mês
calls_per_plan_month['average_call_duration'] = calls_per_plan_month['total_duration'] / calls_per_plan_month['total_calls']

# Visualizar o resultado
print(calls_per_plan_month.head())


# Configurar o estilo do gráfico
sns.set(style="whitegrid")

# Criar um gráfico de linha para visualizar a duração média das chamadas
plt.figure(figsize=(12, 6))
sns.lineplot(data=calls_per_plan_month, x='month', y='average_call_duration', hue='plan', marker='o')

plt.title('Duração Média das Chamadas por Plano e Mês')
plt.xlabel('Mês')
plt.ylabel('Duração Média das Chamadas (minutos)')
plt.legend(title='Plano')
plt.grid(True)
plt.show()
```

## Teste hipóteses estatísticas

```python

# Filtrar as receitas dos planos Ultimate e Surf utilizando a coluna correta
ultimate_revenue = df_combined[df_combined['plan'] == 'ultimate']['monthly_revenue'].dropna()
surf_revenue = df_combined[df_combined['plan'] == 'surf']['monthly_revenue'].dropna()

# Verificar se há variância nos dados
if ultimate_revenue.var() > 0 and surf_revenue.var() > 0:
    # Teste t para amostras independentes
    t_stat, p_value = stats.ttest_ind(ultimate_revenue, surf_revenue, equal_var=False)
    
    # Definir o nível de significância
    alpha = 0.05
    
    # Exibir os resultados
    print(f"Estatística t: {t_stat}")
    print(f"Valor p: {p_value}")

    # Tomar a decisão com base no valor p
    if p_value < alpha:
        print("Rejeita-se a hipótese nula. ")
    else:
        print("Não se rejeita a hipótese nula. ")
else:
    print("Variância zero detectada. O teste t não é aplicável.")

```
